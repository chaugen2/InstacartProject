## Question 4

```python
# Identifying loyal customers, create column
ords_prods_merged['max_order'] = ords_prods_merged.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# Check column has been created
ords_prods_merged.head(15)
```

The output above confirms the column was created.

```python
# Check aggregation
pd.options.display.max_rows = None
```

```python
ords_prods_merged.head(100)
```

The aggregation appears to have been successful.

```python
# Create new column with flags
ords_prods_merged.loc[ords_prods_merged['max_order'] > 40, 'loyalty_flag'] = 'Loyal Customer'
ords_prods_merged.loc[(ords_prods_merged['max_order'] <= 40) & (ords_prods_merged['max_order'] > 10), 'loyalty_flag'] = 'Regular Customer'
ords_prods_merged.loc[ords_prods_merged['max_order'] <= 10, 'loyalty_flag'] = 'New Customer'
```

```python
# Check frequencies of new flag
ords_prods_merged['loyalty_flag'].value_counts(dropna = False)
```

```python
# Check columns
ords_prods_merged[['user_id', 'loyalty_flag', 'order_number']].head(60)
```

## Question 5

```python
# Determine spending habits of different customer types
ords_prods_merged.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']})
```

The above statistics indicate that new customers actually spend more, followed by regular customers, then finally loyal customers. So, while loyal customers order more frequently, they actually spend less per order. There is no difference between the min and max.

## Question 6

```python
# Create new column based on user spending habits
ords_prods_merged['order_avg'] = ords_prods_merged.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# Check column has been created
ords_prods_merged.head(15)
```

```python
# Check aggregation
pd.options.display.max_rows = None
```

```python
ords_prods_merged.head(100)
```

```python
# Create new spending habits column with flags
ords_prods_merged.loc[ords_prods_merged['order_avg'] < 10, 'spender_type'] = 'Low Spender'
ords_prods_merged.loc[ords_prods_merged['order_avg'] >=10, 'spender_type'] = 'High Spender'
```

```python
# Check column has been created
ords_prods_merged.head(15)
```

```python
# Check frequencies of new flag
ords_prods_merged['spender_type'].value_counts(dropna = False)
```

Based on the statistics above, the frequencies make sense.

```python
# Check columns
ords_prods_merged[['order_id','user_id', 'spender_type', 'order_avg']].head(60)
```

## Question 7

```python
# Create new column based on frequency of ordering
ords_prods_merged['median_days_prior_order'] = ords_prods_merged.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# Check column has been created
ords_prods_merged.head(15)
```

```python
# Check aggregation
pd.options.display.max_rows = None
```

```python
ords_prods_merged.head(100)
```

```python
# Create new customer frequency column with flags
ords_prods_merged.loc[ords_prods_merged['median_days_prior_order'] > 20, 'customer_frequency'] = 'Non-Frequent Customer'
ords_prods_merged.loc[(ords_prods_merged['median_days_prior_order'] <= 20) & (ords_prods_merged['median_days_prior_order'] > 10), 'customer_frequency'] = 'Regular Customer'
ords_prods_merged.loc[ords_prods_merged['median_days_prior_order'] <= 10, 'customer_frequency'] = 'Frequent Customer'
```

```python
# Check column has been created
ords_prods_merged.head(15)
```

```python
# Check frequencies of new flag
ords_prods_merged['customer_frequency'].value_counts(dropna = False)
```

```python
# Check all new columns
ords_prods_merged[['order_id','user_id', 'spender_type', 'order_avg', 'customer_frequency','median_days_prior_order']].head(60)
```
